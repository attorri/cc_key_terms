```python
import os

def read_webvtt(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
            return content
    except FileNotFoundError:
        print(f"Error: File not found - {file_path}")
    except Exception as e:
        print(f"An error occurred: {e}")

def main():
    webvtt_file_path = '/Users/christianattorri/Downloads/dummy.vtt' 

    content = read_webvtt(webvtt_file_path)

    if content:
        pcr_found = False
        openshift_found = False
        rhel_found = False

        if 'PCR' in content:
            if not pcr_found:
                print(f"\nPCR (Patient Care Report) is a comprehensive documentation form that records details of pre-hospital patient care, including assessments, interventions, and patient information")
                pcr_found = True
            
        if 'openshift' in content:
            if not openshift_found:
                print(f"\nOpenShift is a tool that helps software developers put their applications on the internet more easily, making it simpler to manage and grow their projects.")
                openshift_found = True
            
        if 'RHEL' in content:
            if not rhel_found:
                print(f"\nRHEL (Red Hat Enterprise Linux) is a computer operating system that provides a stable foundation for running software and applications on servers and business computers.")
                rhel_found = True
        
        try:
            os.remove(webvtt_file_path)
             
        except Exception as e:
            print(f"Couldn't delete file: {e}")
            
if __name__ == "__main__":
    main()
```
